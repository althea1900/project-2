```python
# Dependencies
import requests
import json
import pandas as pd
# Google developer API key
from config import gkey
```

```python
# geocoordinates
target_coordinates = "37.54, -77.45"
target_radius = 10000
target_type = "hospital"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
# print the response url, avoid doing for public github repos in order to avoid exposing key
#print(response.url)
```

```python
# convert response to json
places_data = response.json()

# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))
```

```python
# Print the name and address of the first restaurant that appears
print(places_data["results"][0]["geometry"]["location"]["lat"])
print(places_data["results"][0]["geometry"]["location"]["lng"])
print(places_data["results"][0]["name"])
```

```python
len(places_data["results"])
```

```python
hospital = pd.DataFrame(columns = ["Name", "Longitude", "Latitude", "Type"])
hospital
```

```python
for i in range(len(places_data["results"])):
    hospital = hospital.append({"Name": places_data["results"][i]["name"], "Longitude": places_data["results"][i]["geometry"]["location"]["lng"], "Latitude": places_data["results"][i]["geometry"]["location"]["lat"], "Type": "Hospital"}, ignore_index=True)
```
